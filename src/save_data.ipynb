```python
for pid in all_pids:
    X_dict, Y_dict = load_data_from_pids(
        [pid],
        brain_region=regions[0],
        behavior="wheel_speed",
        # data_type="good_ks",
        data_type="all_ks",
        n_t_bins=n_t_bins,
        t_before=t_before,
        t_after=t_after,
        normalize_input=False
    )
    try:
        X = X_dict[pid]
        Y = Y_dict[pid]

        # Save X and Y
        np.save(f"raw_data/full_data/X_{pid}_{regions[0]}_{n_t_bins}_{t_before}_{t_after}_all-ks.npy", X)
        np.save(f"raw_data/full_data/Y_{pid}_{regions[0]}_{n_t_bins}_{t_before}_{t_after}_all-ks.npy", Y)
    except:
        print(f"Error: {pid}")
```

```python
# Loop through every file in src/raw_data/full_data and print the dimension of the np array in the file along with the name of the file
import os

names_and_shapes = []
for filename in os.listdir("raw_data/full_data"):
    if filename.endswith(".npy"):
        X = np.load(f"raw_data/full_data/{filename}")
        names_and_shapes.append(f"{filename}, with shape {X.shape}")

# sort names_and_shapes
names_and_shapes.sort()
for name_and_shape in names_and_shapes:
    print(name_and_shape)

# save names_and_shapes to a text file
with open("raw_data/full_data/names_and_shapes.txt", "w") as f:
    for name_and_shape in names_and_shapes:
        f.write(f"{name_and_shape}\n")
```
