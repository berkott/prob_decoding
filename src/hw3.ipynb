```python
import numpy as np
import plotly.graph_objs as go
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn import linear_model

N = 14680
D = 24
```

```python
# Load raw_data/small_data_X.npy and raw_data/small_data_Y.npy
X = np.load('raw_data/small_data_X.npy')
Y = np.load('raw_data/small_data_Y.npy')

print(f'Original X shape: {X.shape}, Original Y shape: {Y.shape}')

X = np.transpose(X, (0, 2, 1)).reshape((-1, D))
Y = Y.reshape(-1)

# Print the shape of X and Y using f-strings
print(f'X shape: {X.shape}, Y shape: {Y.shape}')

X_train = X[:11000]
Y_train = Y[:11000]
X_test = X[11000:]
Y_test = Y[11000:]

print(f'X_train shape: {X_train.shape}, Y_train shape: {Y_train.shape}')
print(f'X_test shape: {X_test.shape}, Y_test shape: {Y_test.shape}')
```

```python
# Print the first 20 rows of X_train and Y_train
rows_to_print = 20
for row_to_print in range(rows_to_print):
    print(f'X_train[{row_to_print}]: {X_train[row_to_print]}, Y_train[{row_to_print}]: {Y_train[row_to_print]}')
```

```python
from plotly.subplots import make_subplots

trials = 1
neurons = 24

plot_x = np.arange(40)
plot_speeds = [Y_test[40*trial : 40*(trial + 1)] for trial in range(trials)] 
plot_neurons = [X_test[40*trial : 40*(trial + 1)] for trial in range(trials)] 

fig = make_subplots(specs=[[{"secondary_y": True}]])

for i, plot_speed in enumerate(plot_speeds):
    fig.add_trace(
        go.Scatter(x=plot_x, y=plot_speed, mode='markers+lines', name=f'Session {i+1} speed'),
        secondary_y=True,
    )

for i, plot_neuron in enumerate(plot_neurons):
    for j in range(neurons):
        fig.add_trace(
            go.Scatter(x=plot_x, y=plot_neuron[:, j], mode='markers+lines', name=f'Session {i+1} neuron {j+1}'),
            secondary_y=False,
        )

fig.update_layout(
    xaxis_title="Time bin",
    title="Wheel speed of the first 2 sessions"
)
fig.update_yaxes(title_text="Number of spikes", secondary_y=False)
fig.update_yaxes(title_text="Wheel speed", secondary_y=True)
fig.show()
```

## Enoding

```python
clfs = [linear_model.PoissonRegressor() for _ in range(D)]
X_test_hat = np.zeros((X_test.shape[0], D))

# add a dimension to Y_train and Y_test
Y_train_dim = Y_train[:, np.newaxis]
Y_test_dim = Y_test[:, np.newaxis]
for d in range(D):
    clfs[d].fit(Y_train_dim, X_train[:, d])
    X_test_hat[:, d] = clfs[d].predict(Y_test_dim)

betas = [float(clfs[d].coef_) for d in range(D)]

print(f'MSE: {mean_squared_error(X_test, X_test_hat)}')
print(f"beta: {betas}")
# Plot the betas using plotly
fig = go.Figure()
fig.add_trace(go.Bar(x=np.arange(D), y=betas))
fig.update_layout(
    xaxis_title="Neuron",
    yaxis_title="Value",
    title="Beta for each neuron"
)
fig.show()
```

```python
sessions = 4

plot_z = X_test_hat[:40 * sessions].T
fig = go.Figure(data=go.Heatmap(
                     z=plot_z,
                     colorscale='Viridis'))
fig.update_layout(
    xaxis_title="Time bin",
    yaxis_title="Neuron",
    title="Predicted number of spikes"
)
for i in range(sessions - 1):
    fig.add_shape(type="line",
        x0=40*(i+1), y0=-0.5, x1=40*(i+1), y1=23.5,
        line=dict(color="Red",width=1)
    )
fig.show()

plot_z = X_test[:40 * sessions].T
fig = go.Figure(data=go.Heatmap(
                     z=plot_z,
                     colorscale='Viridis'))
fig.update_layout(
    xaxis_title="Time bin",
    yaxis_title="Neuron",
    title="True number of spikes"
)
for i in range(sessions - 1):
    fig.add_shape(type="line",
        x0=40*(i+1), y0=-0.5, x1=40*(i+1), y1=23.5,
        line=dict(color="Red",width=1)
    )
fig.show()
```

## Decoding

```python
clf = linear_model.Ridge(alpha=.2)
clf.fit(X_train, Y_train)
Y_test_hat = clf.predict(X_test)

print(f'MSE: {mean_squared_error(Y_test, Y_test_hat)}')

# Plot the betas using plotly
fig = go.Figure()
fig.add_trace(go.Bar(x=np.arange(D), y=clf.coef_))
fig.update_layout(
    xaxis_title="Neuron",
    yaxis_title="Value",
    title="Ridge regression beta for each neuron"
)
fig.show()
```
